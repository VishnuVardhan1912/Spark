```python
# 1. Find out how many male and female passengers was onborded.

no_of_male = len(df[df['Sex'] == 'male'])

no_of_female = len(df[df['Sex'] == 'female'])

print({'male': no_of_male, 'female': no_of_male})
```

```python
df['Sex'].value_counts()
```

```python
# 2. how many survived we have

len(df[df['Survived'] == 1])
```

```python
# 3. how many casulaity we have

len(df[df['Survived'] == 0])
```

```python
df['Survived'].value_counts()
```

```python
# 4. What is name of the person who is the eldest one

df[df['Age'] == min(df['Age'])]['Name']
```

```python
# 5. How many passenger do we have in first, second and third class

for i in range(1, 4):
    count = len(df[df['Pclass'] == i])
    print(f"class {i} has {count}")
```

```python
df['Pclass'].value_counts(sort=False)
```

```python
df['Name']
```

```python
# 6. How many persons do we have whose name starts with s

count = 0
for i in df['Name']:
    if i[0] == 's' or i[0] == 'S':
        count += 1
print(f"No of persons starts with s is {count}")
```

```python
# 7.try to create a new column which is summation of "Sibsp" and "parch"

df['Summation'] = df['SibSp'] + df['Parch']
```
